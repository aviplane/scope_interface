```python
import pyvisa
import time
rm = pyvisa.ResourceManager()
res = rm.list_resources()
print(res)
```

```python

inst = rm.open_resource(res[0])
```

```python
inst.query("*IDN?")
```

```python
t_scale = inst.query_ascii_values(":TIM:SCAL?")
v_scale = inst.query_ascii_values(":CHAN1:SCAL?")
v_offset = inst.query_ascii_values(":CHAN1:OFFS?")
t_scale, v_scale, v_offset
inst.query_ascii_values(":TIM:OFFS?")
```

```python
import matplotlib.pyplot as plt
import numpy as np

inst.write(":TIM:SCAL 0.001")
inst.write(":TIM:OFFS 0")

time.sleep(1)
for i in range(1):
    inst.write(":STOP")
    inst.write(':WAV:SOUR CHAN1')
    inst.write(':WAV:MODE norm')
    inst.write(':WAV:FORM byte')
    # inst.write(':WAV:STAR 1')
    # inst.write(':WAV:stop 10000')
    inst.write(":WAV:DATA?")
    data = inst.read_raw()
    wvf_format, wvf_type, wvf_points, count, xinc, xorigin, xref, yinc, yorigin, yref = inst.query_ascii_values(":WAV:PRE?")
    data = np.frombuffer(data, dtype = "uint8",offset = 12)
    fig, ax = plt.subplots()
    print(xinc, xorigin, xref)
    t = np.arange(0, -wvf_points * xinc, -xinc) - xorigin
    ax.plot(t, (data - yref) * yinc - yorigin)
    ax.set_xlabel("t (s)")
    ax.set_ylabel("V (mV)")
    inst.write(":RUN")
    time.sleep(1)
```

```python

fig, ax = plt.subplots()
print(xinc, xorigin, xref)
t = np.arange(0, wvf_points * xinc, xinc) + xorigin
ax.plot(t, (data - yref) * yinc - yorigin)
```
